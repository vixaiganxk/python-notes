```python
from myTask import myTask
import concurrent.futures
from concurrent.futures import ThreadPoolExecutor
import time

task = myTask()
functions = [task.eat_breakfast, task.cooking_food, task.phone_call, task.washing_cloths]


#Executing functions in myTask module
def exec_func(func):
    return func

start_time = time.time()
with ThreadPoolExecutor() as executor:
    for _ in range(10):
        for i in functions:
            results = executor.submit(i)

    #print(results.result())

end_time = time.time()

print(f'Total time taken to complete: {round((end_time-start_time), 4)} second(s)')
```

```python
import threading

threads = []

start_time = time.time()
for _ in range(100):
    t = threading.Thread(target=task.cooking_food)
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()

end_time = time.time()

print(f'Total time taken to complete: {round((end_time-start_time), 4)} second(s)')
```
